```python
# importing pandas as pd
import pandas as pd

# importing numpy as np
import numpy as np

# dictionary of lists
dict = {'First Score':[100, 90, np.nan, 95],
        'Second Score': [30, 45, 56, np.nan],
        'Third Score':[np.nan, 40, 80, 98]}

# creating a dataframe from list
df = pd.DataFrame(dict)

# using isnull() function
df.isnull()
```

```python
# finding full null values
df2 = df.fillna(value = 0)
df2
```

```python
df2.isnull().sum().sum()
```

```python
df3 = df.fillna(value = 5)
df3
```

```python
# # Filling Null Values with the previous value row wise
df4 = df.fillna(method = 'pad')
df4
```

```python
df4.isnull().sum()
```

```python
# Filling Null Values with the next value row wise
df5 = df.fillna(method = 'bfill')
df5
```

```python
# Filling Null Values with the previous value column wise
df6= df.fillna(method = 'pad', axis=1)
df6
```

```python
# Filling Null Values with the previous value column wise
df7= df.fillna(method = 'bfill', axis=1)
df7
```

```python
# Filling different values in null in different columns
df8 = df.fillna({'society': 'abcd', 'balcony': "defg"})
df8
```

```python
# df9 = df.fillna(value=df['balcony'].max())
# df9 = df.fillna(value=df['balcony'].min())

df9 = df.fillna(value=df['balcony'].mean())
df9
```

```python
# drop NAN value method-1
df10 = df.dropna()
df10
```

```python
# drop NAN value method-2, if all is Null value,than all row will be dropped, otherwise not
df11 = df.dropna(how = 'all')
df11
```

```python
# drop NAN value method-2, if any one is Null value,that row will be dropped, otherwise not
df12 = df.dropna(how = 'any')
df12
```

```python
# replace value
import numpy as np
df13= df.replace(to_replace=np.nan, value = 50)
df13
```

```python
df14= df.replace(to_replace=3.0, value = 5.0)
df14
```

```python
# interpolate -> linear , 1,2 3,Nan ->(auto fill with 4),5,6,7
df['balcony'] = df['balcony'].interpolate(method='linear')
df
```

```python
df['balcony'] = df['balcony'].interpolate(method='linear', limit_direction='forward')
df
```
