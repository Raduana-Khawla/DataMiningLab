Summary of Clustering Algorithms:
# **K-Means**: Simple, effective, but requires specifying the number of clusters.
# **Hierarchical Clustering**: Builds a tree of clusters, useful for visualizing with dendrograms.
# **DBSCAN**: Detects clusters of varying shapes, handles noise, and doesn't require the number of clusters.
# **GMM**: Probabilistic clustering, can model clusters with different shapes and sizes.

```python
# Method 1. K-Means Clustering
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt

# Example data
data = np.array([[1, 2], [2, 3], [3, 4], [8, 7], [9, 8], [10, 9]])

# Initialize K-Means with 2 clusters
kmeans = KMeans(n_clusters=2)

# Fit the model
kmeans.fit(data)

# Predict the cluster for each data point
labels = kmeans.predict(data)

# Cluster centers
centers = kmeans.cluster_centers_

print("Labels:", labels)
print("Cluster Centers:", centers)

# Plot the clusters
plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='viridis')
plt.scatter(centers[:, 0], centers[:, 1], c='red', marker='x')
plt.show()
```

```python
# Method 2 : Agglomerative Hierarchical Clustering
from sklearn.cluster import AgglomerativeClustering
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage

# Example data
data = np.array([[1, 2], [2, 3], [3, 4], [8, 7], [9, 8], [10, 9]])

# Perform hierarchical/agglomerative clustering
agg_cluster = AgglomerativeClustering(n_clusters=2)
labels = agg_cluster.fit_predict(data)

# Plot the clusters
plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='viridis')
plt.show()

# Generate the linkage matrix for dendrogram
linked = linkage(data, 'single')

# Plot the dendrogram
plt.figure(figsize=(10, 7))
dendrogram(linked)
plt.show()
```

```python
# Method 3. DBSCAN (Density-Based Spatial Clustering of Applications with Noise)
from sklearn.cluster import DBSCAN
import numpy as np
import matplotlib.pyplot as plt

# Example data
data = np.array([[1, 2], [2, 3], [3, 4], [8, 7], [9, 8], [10, 9], [25, 80]])

# Initialize DBSCAN
dbscan = DBSCAN(eps=3, min_samples=2)

# Fit the model
labels = dbscan.fit_predict(data)

print("Labels:", labels)

# Plot the clusters
plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='viridis')
plt.show()
```

```python
# Method 4. Gaussian Mixture Models (GMM)
from sklearn.mixture import GaussianMixture
import numpy as np
import matplotlib.pyplot as plt

# Example data
data = np.array([[1, 2], [2, 3], [3, 4], [8, 7], [9, 8], [10, 9]])

# Initialize GMM with 2 components
gmm = GaussianMixture(n_components=2)

# Fit the model
gmm.fit(data)

# Predict the cluster for each data point
labels = gmm.predict(data)

# Plot the clusters
plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='viridis')
plt.show()
```

```python
# Method 5. Choosing the Number of Clusters
# Elbow Method for K-Means
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np

# Example data
data = np.array([[1, 2], [2, 3], [3, 4], [8, 7], [9, 8], [10, 9]])

# Determine the optimal number of clusters using the Elbow Method
wcss = []
max_clusters = len(data)  # The maximum number of clusters should be equal to the number of data points

for i in range(1, max_clusters + 1):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(data)
    wcss.append(kmeans.inertia_)

# Plot the results
plt.plot(range(1, max_clusters + 1), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')  # Within-cluster sum of squares
plt.show()
```

```python
#  Silhouette Score
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans

# Example data
data = np.array([[1, 2], [2, 3], [3, 4], [8, 7], [9, 8], [10, 9]])

# Fit K-Means with 2 clusters
kmeans = KMeans(n_clusters=2)
labels = kmeans.fit_predict(data)

# Calculate the silhouette score
score = silhouette_score(data, labels)
print("Silhouette Score:", score)
```

```python
# Method 6. Visualizing Clusters using PCA
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Example data (high-dimensional)
data = np.random.rand(100, 5)

# Apply PCA to reduce to 2D
pca = PCA(n_components=2)
reduced_data = pca.fit_transform(data)

# Fit K-Means with 3 clusters
kmeans = KMeans(n_clusters=3)
labels = kmeans.fit_predict(reduced_data)

# Plot the clusters
plt.scatter(reduced_data[:, 0], reduced_data[:, 1], c=labels, cmap='viridis')
plt.show()
```
